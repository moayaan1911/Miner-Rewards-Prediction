import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
```

importing csv

```python
data= pd.read_csv('blockRewards.csv')
print(data)
```

displaying data

```python
plt.xlabel('timestamp(seconds)')
plt.ylabel('blockRewards(eth)')
plt.scatter(data.timeStamp, data.blockReward, color='red', marker='.')
```

Random Forest Model

```python
model=RandomForestRegressor(n_estimators=100,random_state=42)
model.fit(data.timeStamp.values.reshape(-1,1),data.blockReward.values.ravel())
```

```python
new_timeStamp=np.array([[1466560550]])
predicted_blockReward=model.predict(new_timeStamp)
plt.scatter(data.timeStamp, data.blockReward, label="Data Points")
plt.scatter(new_timeStamp, predicted_blockReward, label="Predicted Future Block Reward")
plt.xlabel('TimeStamp')
plt.ylabel('Block Reward')
plt.legend()
plt.show()
print(predicted_blockReward[0])
```
